```python
# import load_iris function from datasets module
# convention is to import modules instead of sklearn as a whole
from sklearn.datasets import load_iris
```

```python
# save "bunch" object containing iris dataset and its attributes
# the data type is "bunch"
iris = load_iris()
type(iris)
```

```python
# print the iris data
# same data as shown previously
# each row represents each sample
# each column represents the features
print(iris.data)
```

```python
# features of our data
print(iris.feature_names)
```

```python
# integers representing the species of each observation
# 0, 1, and 2 represent different species
print(iris.target)
```

```python
print(iris.target_names)
```

```python
#here the encoding 0 is setosa,1 is versicolor,2 is virginica
```

```python
# check the shape of the features (dimension1 = no of observations,dimension2 = no of features)
print(iris.data.shape)
```

```python
# check the shape of the response (single dimension matching the number of observations)
print(iris.target.shape)
```

```python
# storing feature matrix in "X"
X = iris.data

# storing response vector in "y"
y = iris.target
```
